Let's clean up the data

```python
# import the pandas library and read the data from the csv file
import pandas as pd
df = pd.read_csv('data/PalmerStation_Monthly_Weather.csv')
df.head()
```

```python
# make a new column for the year and month
df['Year'] = pd.to_datetime(df['Date']).dt.year
df['Month'] = pd.to_datetime(df['Date']).dt.month
df.head()
```

```python
# drop the columns that are not needed
df = df[['Date', 'Year', 'Month', 'Mean Temperature (C)']]
df.head()
```

```python
# write the df to a new csv file
df.to_csv('data/PalmerStation_Monthly_Weather_Clean.csv', index=False)
```
